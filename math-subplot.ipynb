```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-4, 4.01, 0.01) # -4から4.01までの範囲で、0.01きざみの値を生成
```

```python
# sin
plt.subplot(231) # 図を何行に分割するか、図を何列に分割するか、このsubplotが左上から数えて何枚目のプロットなのか
# 6(2×3)個に分割して、その中の4枚目のsubplotであることを示す
plt.plot(x, np.sin(x))

# cos
plt.subplot(232)
plt.plot(x, np.cos(x))

# tan
plt.subplot(233)
plt.plot(x, np.tan(x))
plt.ylim(-20, 20)

# sinh
plt.subplot(234)
plt.plot(x, np.sinh(x))

# cosh
plt.subplot(235)
plt.plot(x, np.cosh(x))

# tanh
plt.subplot(236)
plt.plot(x, np.tanh(x))
```
